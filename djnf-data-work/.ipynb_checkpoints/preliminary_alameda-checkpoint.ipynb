```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import contextily as ctx 
import geopandas as gpd 
import os 
import seaborn as sns
from scipy import stats
import warnings
```

```python
#rv = pd.read_csv("alameda_g22_block_registered_voters.csv")
av = pd.read_csv("alameda_g22__block_actual_voters.csv")
dic = pd.read_csv("cvap_dictionary.csv")
cvap = pd.read_csv("alameda_CVAP_2018_2022.csv")
```

```python
warnings.filterwarnings('ignore')

cvap_dic = dic.iloc[14:27]
cvap_dic["DESCRIPTION"] = cvap_dic["DESCRIPTION"].str[31:]
cvap_dic["DESCRIPTION"] = cvap_dic["DESCRIPTION"].str.replace("Not Hispanic or Latino: ", "")
cvap_dic["DESCRIPTION"] = cvap_dic["DESCRIPTION"].str[:-9]
cvap_dic["DESCRIPTION"] = cvap_dic["DESCRIPTION"].str.strip()
cvap_dic
```

```python
dict_race = {}

for i in range(len(cvap_dic["CSV_FIELD"])):
    
    dict_race[cvap_dic["CSV_FIELD"].iloc[i]] = cvap_dic["DESCRIPTION"].iloc[i]
    
dict_race
```

```python
county = "6001"
```

```python
av = av.groupby("tract").agg(sum).reset_index()[["tract", "totreg_r"]]
av
```

```python
cvap["tract"] = cvap["BLOCK20"].astype(str).str[4:].str[:6]
data = cvap[["tract", "CVAP_22"]].copy()
data = data.groupby("tract").sum().reset_index()
data["tract"] = data["tract"].astype(int)
data
```

```python
data = data.merge(av, on="tract")
data
```

```python
data["% CVAP voted"] = data["totreg_r"] / data["CVAP_22"]
data = data.sort_values("% CVAP voted", ascending=False)
data
```

```python
data.sort_values("CVAP_22", ascending=False)
data
```

```python
geo_data = gpd.read_file("tl_2020_06_tract")
geo_data["TRACTCE"] = geo_data["TRACTCE"].astype(int)
geo_data.to_crs("EPSG:4326")
geo_data
```

```python
geo_alameda = geo_data[geo_data["COUNTYFP"] == "001"]
geo_alameda.head(5)
```

```python
geo_alameda.plot()
```

```python
cvap.columns
```

```python
cvap_cvap = cvap[[
                  'COUNTY20', 
                  'BLOCK20', 
                  'CVAP_22', 
                  'NH_CVAP_22', 
                  'NH_Ind_CVAP_22', 
                  'NH_Asn_CVAP_22', 
                  'NH_Blk_CVAP_22', 
                  'NH_Hwn_CVAP_22', 
                  'NH_Wht_CVAP_22', 
                  'NH_IndWht_CVAP_22',
                  'NH_AsnWht_CVAP_22', 
                  'NH_BlkWht_CVAP_22', 
                  'NH_IndBlk_CVAP_22', 
                  'NH_OthMR_CVAP_22', 
                  'Hsp_CVAP_22'
                 ]]

cvap_cvap
```

```python
cvap_cvap.columns = pd.Series(list(cvap_cvap.columns)).replace(dict_race)
cvap_cvap["tract"] = cvap_cvap["BLOCK20"].astype(str).str[4:].str[:6]
cvap_cvap = cvap_cvap.groupby("tract").sum().reset_index()
cvap_cvap
```

```python
geo_alameda
```

```python
races = cvap_cvap.columns[4:]

for i in range(len(races)):
    
    cvap_cvap[races[i] + "%"] = cvap_cvap[races[i]] / cvap_cvap["Total"]
    
cvap_cvap["tract"] = cvap_cvap["tract"].astype(int)
```

```python
data_cvap_cvap = data.merge(cvap_cvap, on="tract").drop("Total", axis=1)
data_cvap_cvap
```

```python
merged = geo_alameda.merge(data_cvap_cvap, how="left", left_on="TRACTCE", right_on="tract")
merged
```

```python
merged_clean = merged[merged["CVAP_22"] > 100]
merged_clean.head(5)
```

```python
merged_clean.to_file('race_voter_turnout_alameda_g.gpkg', driver='GPKG')
```

```python
merged_clean.columns
```

```python
f, ax = plt.subplots(1, 1, figsize= (20,10))
merged_clean.plot('% CVAP voted', ax=ax, cmap='plasma',legend=True)
plt.title("Alameda County census tracts by percent of voting-age population which voted in the 2022 general election")
ax.set_facecolor('k')
plt.savefig('map_alameda_CVAP.png')
```

```python
f, ax = plt.subplots(1, 1, figsize= (20,10))
merged_clean.plot('White Alone%', ax=ax, cmap='plasma',legend=True)
plt.title("")
ax.set_facecolor('k')
plt.savefig('map_alameda_white_%.png')
```

```python
corr_cols = ['% CVAP voted', 'Not Hispanic or Latino%', 'American Indian or Alaska Native Alone%',
       'Asian Alone%', 'Black or African American Alone%',
       'Native Hawaiian or Other Pacific Islander Alone%', 'White Alone%',
       #'American Indian or Alaska Native and White%', 'Asian and White%',
       #'Black or African American and White%',
       #'American Indian or Alaska Native and Black or African American%',
       #'Remainder of Two or More Race Responses%', 
             'Hispanic or Latino%']
```

```python
def corrfunc(x, y, **kws):
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(r),
                xy=(.1, .9),
                xycoords=ax.transAxes)
```

```python
g = sns.PairGrid(merged_clean.dropna()[corr_cols], diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_upper(corrfunc)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
g.savefig("correlations_3.png")
```
