```python
import pandas as pd
import seaborn as sns
from scipy import stats
import geopandas as gpd 
import matplotlib.pyplot as plt 
import matplotlib.ticker as mtick
```

```python
merged = gpd.read_file('race_voter_turnout_alameda_g.gpkg')
merged.crs = 'EPSG:4326'
poverty = pd.read_csv("poverty-level-rates-alameda.csv")
```

```python
poverty.columns = poverty.iloc[0]
poverty = poverty[1:]
poverty = poverty[["Geography", "Geographic Area Name", "Total population for whom poverty status is determined", "Total population below poverty level"]]
poverty["tract"] = poverty["Geographic Area Name"].str[13:].str[:-28]
poverty["tract"] = poverty["tract"].str.replace(".", "")
poverty["tract"] = poverty["tract"].astype(int)
poverty["Total population for whom poverty status is determined"] = poverty["Total population for whom poverty status is determined"].astype(float)
poverty["Total population below poverty level"] = poverty["Total population below poverty level"].astype(float)
poverty
```

```python
poverty["% below poverty level"] = poverty["Total population below poverty level"] / poverty["Total population for whom poverty status is determined"]
poverty = poverty[poverty["Total population for whom poverty status is determined"] > 100]
poverty.sort_values("% below poverty level").head(10)
```

```python
poverty_rate_tracts = poverty[["tract", "% below poverty level"]]
poverty_rate_tracts
```

```python
def standardize_tracts(x):
    
    str_x = str(x)
    len_x = len(str_x)
    
    str_x = str_x + ('0' * (6-len_x))
    
    return int(str_x)
```

```python
poverty_rate_tracts["tract"] = poverty_rate_tracts["tract"].apply(standardize_tracts)
poverty_rate_tracts
```

```python
merged_w_poverty = merged.merge(poverty_rate_tracts, left_on="TRACTCE", right_on="tract")
merged_w_poverty
```

```python
merged_w_poverty.columns
```

```python
def corrfunc(x, y, **kws):
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(r),
                xy=(.1, .9),
                xycoords=ax.transAxes)
```

```python
g = sns.PairGrid(merged_w_poverty[["% below poverty level", "% CVAP voted"]], diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_upper(corrfunc)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
g.savefig("correlations_poverty.png")
```

```python
plt.subplots(1,1,figsize=(10,8))
scat = plt.scatter(merged_w_poverty["% below poverty level"] * 100,
                   merged_w_poverty["% CVAP voted"] * 100,
                   c = merged_w_poverty["White Alone%"] * 100)
plt.xlabel("Percent living below the poverty level", fontsize=12)
plt.ylabel("Percent of the voting-age population which voted", fontsize=12)
plt.title("Racial disparities in voter turnout in Alameda County", fontsize=14)
plt.colorbar(scat)
plt.show()
```
