```python
from scipy import stats
import plotly.express as px
import pandas as pd
```

```python
data = pd.read_csv("table_data.csv")
```

```python
def corr_annotation(x, y):
    pearsonr = stats.pearsonr(x, y)
    return 'r = {:.2f}'.format(pearsonr[0])
```

```python
cols = list(data.columns[6:])

a, b = cols.index('Below poverty level (%)'), cols.index('American Indian or Alaska Native %')
cols[b], cols[a] = cols[a], cols[b]
cols
```

```python
def scatter():
    fig = px.scatter(data, 
                 x='Below poverty level (%)', 
                 y='Voting-age residents who voted in 2022 (%)',
                 width=1000, 
                 height=800) 
    
    fig.add_annotation(dict(text=corr_annotation(data['Below poverty level (%)'], 
                                             data['Voting-age residents who voted in 2022 (%)']),
                        showarrow=False, 
                        yref='paper', 
                        xref='paper',
                        x=0.99, 
                        y=0.95))
    
    buttons = []
    
    for x in cols:
        
        buttons.append(dict(method='update',
                        label=x,
                        args=[{'x': [data[x]],
                               'y': [data['Voting-age residents who voted in 2022 (%)']]},
                              {'xaxis': {'title': x},
                               'yaxis': {'title': 'Voting-age residents who voted in 2022 (%)'},
                               'annotations': [dict(text=corr_annotation(data[x], data['Voting-age residents who voted in 2022 (%)']),
                                                    showarrow=False, 
                                                    yref='paper', xref='paper',
                                                    x=0.99, y=0.95)]}]
                        )
                   )

    fig.update_layout(updatemenus=[dict(buttons=buttons, direction='down', x=0.7, y=1.1)])
    fig.show()
```

```python
scatter()
```
