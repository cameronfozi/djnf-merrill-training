## Exploring the data

Use the scatter plot below to see the correlation between the the percentage of voting-age residents who voted in Bay Area census tracts and nine other demographic indicators: the census tract's percent make-up of White, American Indian or Alaska Native, Asian, Black, Native Hawaiian or other Pacific Islander, Hispanic or Latino and mixed race residents; the percent of residents living below the poverty level and the median age of the census tract.

```python
from ipynb.fs.full.plotly_function import scatter
import plotly.io as pio
from scipy import stats
import plotly.express as px
import pandas as pd
pio.renderers.default = "notebook_connected"
scatter()
```

On the top right, the `r = ...` section refers to the correlation coefficient between the voter turnout percentage and the other indicator. Of all demographic indicators, the strongest indicator of a census tract's voter turnout the percent make-up of the White population. As the White population of a Bay Area census tract goes up, we see an incline in voter turnout.

The next most powerful indicators are the median age and the percent make-up of Hispanic or Latino residents. As age rises, so does voter turnout. But more Hispanic or Latino census tracts tend to show decreased voter turnout.

The correlation coefficient, or r, is a measure of how correlated two variables - collections of data - are. 
- If two variables have a correlation coefficient close to 0, they are less correlated. 
- If two variables have a correlation coefficient close to 1, they have a strong positive correlation. 
- If two variables have a correlation coefficient close to -1, they have a strong negative correlation. 

Here is how correlated the demographic indicators included in this analysis are with voter turnout, in order:
- White (%), r = 0.71
- Median age, r = 0.61
- Hispanic or Latino (%) = -0.61
- Below poverty level (%), r = -0.52
- Black (%), r = -0.37
- Native Hawaiian or other Pacific Islander (%), r = -0.21
- Asian (%), r = -0.13
- American Indian or Alaska Native (%), -0.11
- Mixed race (%), r = 0.01

## How else could this project leverage data?

**1. Time series analysis** <br>
Analyzing voter turnout across general elections may  reveal trends with respect to demographic groups voting trends over time. <br> <br>
**2. Using the census block as the primary unit of spatial analysis**<br>
While this analysis had originally planned to use the census block as a unit of analysis, there were about 700 census blocks of Alameda Countys approximately 18,500 for which there were no residents or the number of voters exceeded the number of voting-age residents, as recorded by the 2022 American Community Survey. Another avenue of exploration could use data from the 2020 decennial census, which are potentially more accurate, and compare them to voter turnout rates from the 2020 general election. <br> <br>
**3. Average distance from a voting precinct**<br>
California has voting-day registration. Say, if someone was not registered to vote the day of an election, how easy would it be for them to register and vote that day? One potential metric for this would be calculating the average distance as the crow flies between all addresses in a census tract and the location of the voting precinct. This analysis  may assume the distance from one's home to a voting precinct is a factor in people's willingness to vote. However, it appears many urban, walkable areas still have abysmal voter turnout, and sprawling, suburban areas have high voter turnout.
